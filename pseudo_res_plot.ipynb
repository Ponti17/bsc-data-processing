```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib as mpl

# Border width
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['font.weight'] = 'regular'
legend_properties = {'weight':'regular'}
mpl.rcParams['font.size'] = 11

dir = "report_data"
filename = "pseudo_res.csv"
data = pd.read_csv(os.path.join(dir, filename))

fig, ax1 = plt.subplots()
fig.set_size_inches(5, 4)

V   = data["RA X"]
RA  = data["RA Y"]
RB  = data["RB Y"]
RC  = data["RC Y"]

ax1.plot(V, RA, linewidth = 2, color="blue", label="A", alpha=0.75)
ax1.plot(V, RB, linewidth = 2, color="red", label="B", alpha=0.75)
ax1.plot(V, RC, linewidth = 2, color="green", label="C", alpha=0.75)

ax1.ticklabel_format(axis='y', style='sci', scilimits=(9, 9)) 
ax1.grid(linestyle='dotted')
ax1.set_xlabel('Voltage Swing [V]')
ax1.set_ylabel('Resistance [$\Omega$]')
ax1.set_xticks([0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9])
ax1.set_yticks([0, 100e9, 200e9, 300e9, 400e9])

ax1.legend()
ax1.set_box_aspect(1)
ax1.set_xlim(0.3, 0.9)
ax1.set_ylim(0, 400e9)
fig.tight_layout()

filename = "mosres_abc.pdf"
path = os.path.join("../report/imgs", filename)
overwrite = False
if not os.path.isfile(path) or overwrite:
    print("Saving to", path)
    plt.savefig(path)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib as mpl

# Border width
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['font.weight'] = 'regular'
legend_properties = {'weight':'regular'}
mpl.rcParams['font.size'] = 11

dir = "report_data"
filename = "pseudo_res.csv"
data = pd.read_csv(os.path.join(dir, filename))

fig, ax1 = plt.subplots()
fig.set_size_inches(8, 4)

V           = data["RA X"]
R_active    = data["R_active Y"]
RF          = data["RF Y"]

ax1.plot(V, RF, linewidth = 2, color="green", label="F", alpha=0.75)
ax1.plot(V, R_active, linewidth = 2, color="blue", label="Active", alpha=0.75)

ax1.ticklabel_format(axis='y', style='sci', scilimits=(9, 9)) 
ax1.grid(linestyle='dotted')
ax1.set_xlabel('Voltage Swing [V]')
ax1.set_ylabel('Resistance [$\Omega$]')
ax1.set_xticks([0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9])
ax1.set_yticks([0, 100e9, 200e9, 300e9, 400e9])

ax1.legend()
ax1.set_box_aspect(1)
ax1.set_xlim(0.3, 0.9)
fig.tight_layout()

filename = "mosres_act.pdf"
path = os.path.join("../report/imgs", filename)
overwrite = True
if not os.path.isfile(path) or overwrite:
    print("Saving to", path)
    plt.savefig(path)
```
