```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib as mpl

# Border width
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['font.weight'] = 'regular'
legend_properties = {'weight':'regular'}
mpl.rcParams['font.size'] = 11

# /Vout X,/Vout Y,/I52/anal_in_p X,/I52/anal_in_p Y,/I52/anal_in_n X,/I52/anal_in_n Y
dir = "report_data"
filename = "ADC_tran.pkl"
data = pd.read_pickle(os.path.join(dir, filename))

fig, ax = plt.subplots()
fig.set_size_inches(5, 4)

t = data["/Vout X"]
out = data["/Vout Y"]
vn = data["/I52/anal_in_n Y"]
vp = data["/I52/anal_in_p Y"]
vin = (vp - vn)*1e3

ax.plot(t, out, label="Output", color='black')
ax.plot(t, out, label="Output", color='black')
ax2 = ax.twinx()
ax2.plot(t, vin, label="Input", color='blue')

periods = 10
start = periods*(1/225)
end = start + (1/225)
ax.set_xlim(start, end)
ax.set_ylim(-0.1, 0.1)

ax.grid(linestyle='dotted')
ax.set_xlabel('Frequency [Hz]')
ax.set_ylabel('Magnitude [dB]')
ax.set_box_aspect(1)
fig.tight_layout()

filename = "twostage_lp_ac_cl.pdf"
path = os.path.join("../report/imgs", filename)
overwrite = False
if not os.path.isfile(path) or overwrite:
    print("Saving to", path)
    plt.savefig(path)
```
