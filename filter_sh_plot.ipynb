```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib.ticker

dir = "amplifiers"
filename = "sample_hold.csv"
data = pd.read_csv(os.path.join(dir, filename))

# set theme and figure size
sns.set_theme(style="darkgrid", palette='tab10', rc={'figure.figsize':(15,6)})
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)

t           = data["v /net1; tran (V) X"]
clk         = data["v /clk; tran (V) Y"]
sig         = data["v /net1; tran (V) Y"]
out         = data["v /VOUT; tran (V) Y"]
titles      = ["$OL Gain$", "OL Phase"]

ax.plot(t, clk, label="CLK", color="blue")
ax.plot(t, sig, label="Input", color="red")
ax.plot(t, out, label="Output", color="green")

fig.tight_layout()

def generate_log_scale(decades):
    scale = []
    for i in range(0, decades):
        scale.append(10**i)
    return scale

ax.grid(True)
ax.legend(loc="upper right")
plt.title('Full Input Swing Folded Cascode')
plt.xlabel('Time (s)')
plt.ylabel('Degrees')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib.ticker

dir = "amplifiers"
filename = "filter.csv"
data = pd.read_csv(os.path.join(dir, filename))

# set theme and figure size
sns.set_theme(style="darkgrid", palette='tab10', rc={'figure.figsize':(15,6)})
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)

freq        = data["v /VOUT; pac dB20(V) (harmonic=0) X"]
mag         = data["v /VOUT; pac dB20(V) (harmonic=0) Y"]

ax.semilogx(freq, mag, color="blue", label="Magnitude [dB]")

fig.tight_layout()

def generate_log_scale(decades):
    scale = []
    for i in range(0, decades):
        scale.append(10**i)
    return scale

ax.set_xscale("log")
ax.grid(True)
# ax.legend(loc="upper right")
plt.title('Full Input Swing Folded Cascode')
plt.xticks(generate_log_scale(6))
plt.xlabel('Frequency')
plt.ylabel('Degrees')
plt.show()
```
