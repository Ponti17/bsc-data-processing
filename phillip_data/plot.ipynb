```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib as mpl

# Border width
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['font.weight'] = 'regular'
legend_properties = {'weight':'regular'}
mpl.rcParams['font.size'] = 11

averages = {
    "CS": [],
    "Lewis": [],
    "LewisALT": []
}
comparators = ["CS", "Lewis", "LewisALT"]
voltages = ["400m", "500m", "600m", "700m", "800m", "900m", "1"]
voltages_float = [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
for comparator in comparators:
    for voltage in voltages:
        dir = "offsetVoltage"
        filename = "OffsetVoltage{0}{1}VCM.csv".format(comparator, voltage)
        data = pd.read_csv(os.path.join(dir, filename))
        x = data.filter(like="X").to_numpy().flatten()
        avg = np.mean(x)
        averages[comparator].append(avg)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(8, 4)

lns1 = ax1.plot(voltages_float, averages["CS"], linewidth = 2, color="red", label="(10)")
lns2 = ax1.plot(voltages_float, averages["LewisALT"], linewidth = 2, color="blue", label="(11)")
#ax2 = ax1.twinx()
lns3 = ax2.plot(voltages_float, averages["Lewis"], linewidth = 2, color="black", label="(9)")

ax1.set_box_aspect(1)
ax2.set_box_aspect(1)
ax1.grid(linestyle='dotted')
ax2.grid(linestyle='dotted')
ax1.set_xlabel('Common Mode Voltage [V]')
ax2.set_xlabel('Common Mode Voltage [V]')
ax1.set_ylabel('Offset Voltage [V]')
ax1.ticklabel_format(axis='y', style='sci', scilimits=(-3, -3))
ax2.ticklabel_format(axis='y', style='sci', scilimits=(-3, -3))
ax1.set_ylim(-0.1*1e-3, 0.4*1e-3)
ax2.set_ylim(-10*1e-3, 40*1e-3)
ax1.legend()
ax2.legend()
#ax2.ticklabel_format(axis='y', style='sci', scilimits=(-3, -3))

fig.tight_layout()

filename = "comparator_mean.pdf"
path = os.path.join("./imgs", filename)
overwrite = True
if not os.path.isfile(path) or overwrite:
    print("Saving to", path)
    plt.savefig(path)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import os
import matplotlib as mpl

# Border width
mpl.rcParams['axes.linewidth'] = 1
mpl.rcParams['font.weight'] = 'regular'
legend_properties = {'weight':'regular'}
mpl.rcParams['font.size'] = 11

averages = {
    "CS": [],
    "Lewis": [],
    "LewisALT": []
}
comparators = ["CS", "Lewis", "LewisALT"]
voltages = ["400m", "500m", "600m", "700m", "800m", "900m", "1"]
voltages_float = [0.4, 0.5, 0.6, 0.7, 0.8, 0.9, 1.0]
for comparator in comparators:
    for voltage in voltages:
        dir = "offsetVoltage"
        filename = "OffsetVoltage{0}{1}VCM.csv".format(comparator, voltage)
        data = pd.read_csv(os.path.join(dir, filename))
        x = data.filter(like="X").to_numpy().flatten()
        avg = np.std(x)
        averages[comparator].append(avg)

fig, (ax1, ax2) = plt.subplots(1, 2)
fig.set_size_inches(8, 4)

lns1 = ax1.plot(voltages_float, averages["CS"], linewidth = 2, color="red", label="(10)")
lns2 = ax1.plot(voltages_float, averages["LewisALT"], linewidth = 2, color="blue", label="(11)")
#ax2 = ax1.twinx()
lns3 = ax2.plot(voltages_float, averages["Lewis"], linewidth = 2, color="black", label="(9)")

ax1.set_box_aspect(1)
ax2.set_box_aspect(1)
ax1.grid(linestyle='dotted')
ax2.grid(linestyle='dotted')
ax1.set_xlabel('Common Mode Voltage [V]')
ax2.set_xlabel('Common Mode Voltage [V]')
ax1.set_ylabel('Offset Voltage [V]')
ax1.ticklabel_format(axis='y', style='sci', scilimits=(-3, -3))
ax2.ticklabel_format(axis='y', style='sci', scilimits=(-3, -3))
ax1.set_ylim(1*1e-3, 6*1e-3)
ax2.set_ylim(-1*1e-3, 180*1e-3)
ax1.legend()
ax2.legend()

fig.tight_layout()

filename = "comparator_std.pdf"
path = os.path.join("./imgs", filename)
overwrite = True
if not os.path.isfile(path) or overwrite:
    print("Saving to", path)
    plt.savefig(path)
```
