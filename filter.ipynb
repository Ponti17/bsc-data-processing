```python
import scipy as sp
import matplotlib.pyplot as plt
import numpy as np

sys = sp.signal.TransferFunction([(2*np.pi*1000)**2], [1, (2*np.pi*1000)/0.707, (2*np.pi*1000)**2])
w, mag, phase = sp.signal.bode(sys)

fig, ax = plt.subplots()
fig.set_size_inches(12, 8)

ax.set_xlabel('Frequency [rad/s]')
ax.set_ylabel('Magnitude [dB]')
ax.semilogx(w, mag)

ax2 = ax.twinx()
ax2.semilogx(w, phase, color="red")
ax2.set_ylabel('Phase [deg]')
ax.grid()

fig.tight_layout()
```

```python
import numpy as np

def closest_index(lst, K):
    return min(range(len(lst)), key=lambda i: abs(lst[i] - K))

K = 2 * np.pi * 10000

closest_index_to_K = closest_index(w, K)
print(closest_index_to_K)
```
