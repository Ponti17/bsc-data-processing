```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = "nmos-gmid-idw-2-20u.csv"
data = pd.read_csv(file_path)

# set theme and figure size
sns.set_theme(style="darkgrid", palette='tab10', rc={'figure.figsize':(15,6)})
fig, ax = plt.subplots()

X_data = [title for title in data.columns if 'X' in title]
Y_data = [title for title in data.columns if 'Y' in title]

print(X_data)

for i in range(0, len(X_data)):
    if i > 3:
        if i % 3 == 0:
            sns.lineplot(x=X_data[i], y=Y_data[i], data=data, ax=ax, label='L = {}u'.format(i*2+2))
    else:
        sns.lineplot(x=X_data[i], y=Y_data[i], data=data, ax=ax, label='L = {}u'.format(i*2+2))

plt.xscale('log')
plt.title('$g_m/I_D$ Versus $I_N$')
plt.xlabel('Reference Current $I_N = I_D/W$')
plt.ylabel('$g_m/I_D$')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

file_path = "gain-cmrr.csv"
data = pd.read_csv(file_path)

# set theme and figure size
sns.set_theme(style="darkgrid", palette='tab10', rc={'figure.figsize':(15,6)})
fig, ax = plt.subplots()

X_data = [title for title in data.columns if 'X' in title]
Y_data = [title for title in data.columns if 'Y' in title]
titles = ["$A_V$", "$A_0$", "CMRR"]

ymin = -20
ymax = 90
for i in range(0, len(X_data)):
    sns.lineplot(x=X_data[i], y=Y_data[i], data=data, ax=ax, label=titles[i])
    max_gain = np.max(data[Y_data[i]])
    for j in range(0, len(data[Y_data[i]])):
        if data[Y_data[i]][j] < max_gain - 3:
            ax.axvline(x=data[X_data[i]][j],ymax=(data[Y_data[i]][j] + np.abs(ymin)) / (ymax + np.abs(ymin)), color='black')
            print("3dB {0}: {1}".format(X_data[i], data[X_data[i]][j]))
            break
ax.set_ybound(ymin, ymax)

def generate_log_scale(decades):
    scale = []
    for i in range(0, decades):
        scale.append(10**i)
    return scale

plt.xscale('log')
plt.xticks(generate_log_scale(11))
plt.title('OTA Performance')
plt.xlabel('Frequency')
plt.ylabel('dB20')
plt.show()
```
